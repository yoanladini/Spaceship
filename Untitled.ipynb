```python
train=pd.read_csv('train.csv')
test=pd.read_csv('test.csv')
```

```python
train['Name']=train['Name'].fillna('JOHN DOE')
test['Name']=test['Name'].fillna('JOHN DOE')
```

```python
train.loc[train['CryoSleep'] == True, ['RoomService', 'FoodCourt', 'ShoppingMall', 'Spa', 'VRDeck']] = 0.0
train.loc[train['CryoSleep'] == True, 'VIP'] = False

test.loc[test['CryoSleep'] == True, ['RoomService', 'FoodCourt', 'ShoppingMall', 'Spa', 'VRDeck']] = 0.0
test.loc[test['CryoSleep'] == True, 'VIP'] = False
```

```python
train.loc[train[['RoomService', 'FoodCourt', 'ShoppingMall', 'Spa', 'VRDeck']].sum(axis=1)==0.0,['CryoSleep']]=True
test.loc[test[['RoomService', 'FoodCourt', 'ShoppingMall', 'Spa', 'VRDeck']].sum(axis=1)==0.0,['CryoSleep']]=True
```

```python
train['CryoSleep'] = train['CryoSleep'].fillna(False)
test['CryoSleep'] = test['CryoSleep'].fillna(False)
```

```python
train['VIP'].value_counts(dropna=False)
```

```python
train[['G_passenger','N_passenger']]=train['PassengerId'].str.split('_',expand=True).astype('int')
test[['G_passenger','N_passenger']]=test['PassengerId'].str.split('_',expand=True).astype('int')
```

```python
train.set_index('PassengerId',inplace=True)
test.set_index('PassengerId',inplace=True)
```

```python
train.Cabin.fillna('T/0/P',inplace=True)
test.Cabin.fillna('T/0/P',inplace=True)
```

```python
train[['Deck','Num','Side']]=train['Cabin'].str.split('/',expand=True)
test[['Deck','Num','Side']]=test['Cabin'].str.split('/',expand=True)
```

```python
train.drop(['Name','Cabin'],axis=1,inplace=True)
test.drop(['Name','Cabin'],axis=1,inplace=True)
```

```python
#Fill Na with Most Freq strategy
imp = SimpleImputer(strategy="most_frequent")
imp.fit_transform(train[['HomePlanet']])

train['HomePlanet']=imp.fit_transform(train[['HomePlanet']])

imp = SimpleImputer(strategy="most_frequent")
imp.fit_transform(train[['Destination']])

train['Destination']=imp.fit_transform(train[['Destination']])

imp = SimpleImputer(strategy="most_frequent")
imp.fit_transform(train[['VIP']])

train['VIP']=imp.fit_transform(train[['VIP']])



#Fill Na with Most Freq strategy
imp = SimpleImputer(strategy="most_frequent")
imp.fit_transform(test[['HomePlanet']])

test['HomePlanet']=imp.fit_transform(test[['HomePlanet']])

imp = SimpleImputer(strategy="most_frequent")
imp.fit_transform(test[['Destination']])

test['Destination']=imp.fit_transform(test[['Destination']])

imp = SimpleImputer(strategy="most_frequent")
imp.fit_transform(test[['VIP']])

test['VIP']=imp.fit_transform(test[['VIP']])

```

```python
train['VIP'].value_counts(dropna=False)
```

```python
imp = SimpleImputer(strategy="median")
imp.fit_transform(train[['Age']])

train['Age']=imp.fit_transform(train[['Age']])

imp = SimpleImputer(strategy="median")
imp.fit_transform(test[['Age']])

test['Age']=imp.fit_transform(test[['Age']])
```

```python
pd.DataFrame(test.isnull().sum(), columns = ['missing_count']).assign(missing_pct = lambda x: x['missing_count']/test.shape[0]*100) . T
```

```python
train[['RoomService', 'FoodCourt', 'ShoppingMall', 'Spa', 'VRDeck']]=train[['RoomService', 'FoodCourt', 'ShoppingMall', 'Spa', 'VRDeck']].fillna(0.0)
test[['RoomService', 'FoodCourt', 'ShoppingMall', 'Spa', 'VRDeck']]=test[['RoomService', 'FoodCourt', 'ShoppingMall', 'Spa', 'VRDeck']].fillna(0.0)
```

```python
train.dtypes
```

```python
train.CryoSleep=train.CryoSleep.astype('object')
test.CryoSleep=test.CryoSleep.astype('object')
train.Transported=train.Transported.astype('object')
```

```python
categorical_cols= ['HomePlanet','CryoSleep','Destination','VIP','Deck','Side','Num']
```

```python
train['Transported']=train['Transported'].replace({True:1,False:0})
```

```python
x=train.drop('Transported',axis=1)
y=train.Transported
```

```python
from catboost import CatBoostClassifier
```

```python
x.dtypes
```

```python
model.fit(x,y,categorical_cols)
```

```python
from sklearn.metrics import accuracy_score
pred=model.predict(x)
print(accuracy_score(y.values,pred))
```
